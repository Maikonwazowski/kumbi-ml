```python
import pandas as pd

data = pd.read_csv('customers.csv')
```

```python
data.head(10)
```

```python
# removing some columns that not make sense 
data.drop(columns=["CUST_ID", "TENURE"], inplace=True)
data.head(10)
```

```python
missing = data.isna().sum()
print(missing)
```

```python
# Eliminando os dados nulos
data.fillna(data.median(), inplace=True)
missing = data.isna().sum()
print(missing)
```

```python
data.head()
```

```python
# Normalizando os dados
from sklearn.preprocessing import Normalizer
values = Normalizer().fit_transform(data.values)
print(values)
```

```python
# Clusterizando

from sklearn.cluster import KMeans

# n_init = número de vezes que não se repete para garantir a confiabilidade
# max_iter = número máximo de iterações
kmeans = KMeans(n_clusters=3, n_init=10, max_iter=300)
y_pred = kmeans.fit_predict(values)

print(y_pred)
```

```python

# Calculando o silhouette


# varia de [-1, 1], se for positivo adotamos com um bom valor
from sklearn import metrics

labels = kmeans.labels_
silhouette = metrics.silhouette_score(values, labels, metric='euclidean')
print(silhouette)
```

```python
# Calculando davies-Bouldin

dbs = metrics.davies_bouldin_score(values, labels)
dbs
```

```python
# Calculando calinski

calinski = metrics.calinski_harabasz_score(values, labels) 
calinski
```

```python
def clustering_algorithm(n_clusters, dataset):
    kmeans = KMeans(n_clusters=n_clusters, n_init=10, max_iter=300)
    y_pred = kmeans.fit_predict(dataset)
    s = metrics.silhouette_score(dataset, labels, metric='euclidean')
    callinski = metrics.calinski_harabasz_score(dataset, labels)
    dbs = metrics.davies_bouldin_score(dataset, labels)
    return s, dbs, callinski
```

```python
s1, dbs1, callinski1 = clustering_algorithm(3, values)
print('silhoute: ', s1 ,'  -  dbs: ',dbs1,'  -  callinski',callinski1)
```

```python
s2, dbs2, callinski2 = clustering_algorithm(5, values)
print('silhoute: ', s2 ,'  -  dbs: ',dbs2,'  -  callinski',callinski2)
```

```python
s3, dbs3, callinski3 = clustering_algorithm(20, values)
print('silhoute: ', s3 ,'  -  dbs: ',dbs3,'  -  callinski',callinski3)
```
